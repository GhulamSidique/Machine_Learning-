### Normalization is a technique in which we rescale data in a certain range say 0 to 1

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# for normalization
from sklearn.preprocessing import MinMaxScaler
```

```python
# load dataset
d_set=sns.load_dataset('titanic')
d_set.head(5)
```

```python
d_set.describe()
```

```python
# making use of the fare column 
# visualisation
sns.distplot(d_set['fare'])
plt.show()
```

```python
# now normalizing the data using min-max scalling technique
# min-max normalization
# Xnew = (Xi - min(X)) / (max(X) - min(x))
# but here we will make use of the sklearn

scale=MinMaxScaler()
scale.fit(d_set[['fare']])
```

```python
# now transfomation
d_set['scaled fare'] =pd.DataFrame(scale.transform(d_set[['fare']]), columns=['x']) 
```

```python
d_set
```

```python
# now visualizing the normalized data
sns.distplot(d_set['scaled fare'])
plt.show()
```

# Result:
### As it can be seen that the data is normalized to 0 and 1 range
