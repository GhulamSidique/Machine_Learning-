### Multiple linear regression is applied when we have more than one input column

```python
# import libraries

# for data 
import pandas as pd
import numpy as np

# for visualisation
import matplotlib.pyplot as plt
import seaborn as sns

# for train and test split
from sklearn.model_selection import train_test_split

# for model
from sklearn.linear_model import LinearRegression
```

```python
# read the dataset
d_set=pd.read_csv('salary.csv')
d_set
```

```python
# data description
d_set.describe()
```

```python
# checking vislaully the outliers the columns
plt.figure(figsize=(15, 5))

plt.subplot(1,3, 1)
sns.boxplot(d_set['age'])
plt.title("Age")

plt.subplot(1,3,2)
sns.boxplot(d_set['experience'])
plt.title("Exp")

plt.subplot(1,3,3)
sns.boxplot(d_set['income'])
plt.title("salary")

plt.show()
```

### result: only few negligible outliers are there in the exp and salary column 

### checking the pattren of the data to choose the model for training

```python
# data visualistaion
plt.figure(figsize=(15, 5))

plt.subplot(1,2,1)
sns.scatterplot(x=d_set['income'], y=d_set['experience'], data=d_set)
plt.title("income vs exp")

plt.subplot(1,2,2)
sns.scatterplot(x=d_set['age'], y=d_set['income'], data=d_set)
plt.title("income vs age")


plt.show()
```

```python
# using heatmap
sns.heatmap(data=d_set.corr(), annot=True)
plt.show()
```

### result: from the graphs it can be seen the relationship is linear with the output column. Therefore, we may make use of linear reg here too.
### The heatmap also shows the relationship between the values of columns using correlation 

```python
# separate the input and output data
X=d_set.drop(columns=['income'])
y=d_set['income']
```

```python
# split the data
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.1)
```

```python
# model training
lr=LinearRegression()

# fit the training data
lr.fit(X_train, y_train)
```

```python
# predict the values
# from the dataset the salary ate age 25 with experience 1 is 30450
lr.predict([[25, 1]])
```

### Result: the predicted value is almost close to the actaual value

```python
# accuracy score
lr.score(X_test, y_test)
```

### result: according to the accuracy score the model performs 90% well
