### in this notebook we will be wroking with decision tree with multiple regression dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
```

```python
# load the dataset
d_set=pd.read_csv("multiple_linear_regression_dataset.csv")
d_set
```

```python
# pairplot
sns.pairplot(data=d_set)
plt.show()
```

### result: from above graph it can be seen that the data points are almost linear separable

```python
# separating the input and output data
x=d_set.iloc[:,:-1]
y=d_set['income']

# split the data
xtrain, xtest, ytrain, ytest=train_test_split(x, y, test_size=0.2)
```

```python
# since we are dealing with regression therefore our model will be regressor
dr=DecisionTreeRegressor()

# fit the data
dr.fit(xtrain, ytrain)
```

```python
# score
print("training--> ", dr.score(xtrain, ytrain)*100,"\n", "testing--> ", dr.score(xtest, ytest)*100)
```

```python
# plot tree
from sklearn.tree import plot_tree
plot_tree(dr)
plt.show()
```
