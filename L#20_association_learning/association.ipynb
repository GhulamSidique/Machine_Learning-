### association learning is used to discover interesting relationships or patterns in large dataset
### often applied to transactional data where items are bought or used together.
### works on the if and else statement such as if A the B. if element is called antecedent and then statement is called consedent.
### 3 techniques are used that are Support, Confidence and Lift
## 1) Support:
### the first step and is used to find the frequent bought items and is the straughtforwar calculation based on frequency.
### Support(A) = Transactions(A) / Total Transactions
## 2) Confidence:
### the second step that tells us how confident we can be that an item will be pruchased given that another item has been purchased.
### Conf (X==>Y) = P(Y|X) = (supp(X intersaction Y)) / (supp(X)) = (no of transactions containing X and Y) / (no of transactions containing X)
## Lift:
### lift (X==>Y) = (supp(X intersaction Y)) / (supp(X) * supp(Y))
### lift > 1: means that the two items are more likely to be bought together and hence keep both items together
### lift < 1: means that the two items are more likely to be bought separately and hence no need to keep both items together but shoould be placed close to each other.
### lift = 1: means that there is no association between the two items. 

### 3 types of association learning are as under
## 1) Apriori algo:
### Algorithm Steps:
#### Step 1 (k=1): Create a table (C1) with the support count of each item. Remove items with support counts below a minimum threshold.
#### Step 2 (k=2): Generate candidate sets (C2) by joining frequent itemsets from the previous step. Check subsets for frequency and prune accordingly.
#### Step 3 (k=3+): Continue generating candidate sets (C3, C4, etc.) and pruning based on the Apriori property.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
d_set=pd.read_csv("groceries - groceries.csv")
d_set.head()
```

### as it can be seen that there are many nan values and we need to remove them

### the type of the nan vaulues is always float and that of categroical data is a string and based on it we will separfate our data

```python
# nan value type check
type(d_set['Item 5'][0])
```

```python
# now we have to create a list of market in which all items other than nan will be kept
market=[]

# loop around the number of rows
for i in range(0, d_set.shape[0]): # .shape will run till the last row value

    # now we have to separate the nan value and keep all other values in the customer's list because those items are sold to them
    customers=[]

    # now loop around the columns to separate the nan type values
    for j in d_set.columns:

        # condition
        if type(d_set[j][i])==str:
            # append the values in customer's list
            customers.append(d_set[j][i])

    # finally append the customer data in te market list
    market.append(customers)
# the market list will contain all the nested list of the items that are bought by the customers
# since the market is a nested list and we need to convert it into a single numpy array as follows
l=[]

# loop around the market values
for i in market:
    # loop around the i
    for j in i:
        l.append(j)
```

```python
# now count the frequency that will show the number of sold items in descending order
# import library
import collections

# count the items
counted=collections.Counter(l)
counted
```

```python
# convert the counted values into dataframe
# fisrt make a dictionary that will hold the item name and corresponding selling values
d={"Item name":counted.keys(), "values":counted.values()}

# now dataframe
df=pd.DataFrame(d).sort_values(by=['values'], ascending=False)
df.head()
```

```python
# now model
from mlxtend.preprocessing.transactionencoder import TransactionEncoder

# model
te=TransactionEncoder()

# fit the market data
te.fit(market)

# make dataframe of it
m_df=pd.DataFrame(te.transform(market), columns=te.columns_)
m_df.head()
```

### result: above dataframe shows the relationship between two selling items. The False means these two items are not sold together

```python
# now performing apriori algo
from mlxtend.frequent_patterns import apriori

# model
apriori(m_df, min_support=0.05, use_colnames=True, max_len=3).sort_values(by=['support'], ascending=False)
```

### result: from above values there are 3 combinations of 2 items that are sold together.
### but the support value for (other vegetables, whole milk) = 0.074835 is greater than theat of other two combinations. therefor, we will make use of (other vegetables, whole milk) for the next step
